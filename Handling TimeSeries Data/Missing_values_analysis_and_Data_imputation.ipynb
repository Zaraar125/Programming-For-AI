### 1. Drop rows or columns that have a missing value

```python
df_spy.dropna()     #Drop Rows
```

```python
df_spy.dropna(axis=1)
```

```python
df_spy.dropna(how='all') #Drop rows or columns that only have missing values (full of missing value So, nothing is deleted)
```

#### Drop rows or columns based on a threshold value

```python
df_spy.dropna(thresh=5)  # “thresh=5” means that the rows that have at least 4 non-missing values will be kept. The other ones will be dropped.
```

#### Drop based on a particular subset of columns

```python
df_spy.dropna(subset=["Open","Low"]) #rop the rows that have a missing value in Open and Low
```

#### Fill with a constant value

```python
values={"Open":120.0,"Close":121.0}
df_spy.fillna(value=values)
```

#### Fill with an aggregated value

```python
df_spy["High"].fillna(df_spy["High"].mean())
```

#### Replace with the previous or next value

```python
df_spy.fillna(method='bfill') # Replace with the next value
```

```python
df_spy
```

```python
df_spy.fillna(method='pad')  # We use method = 'pad’ for taking values from the previous row.
```

Pandas `dataframe.ffill()` function is used to fill the missing value in the dataframe. `ffill` stands for ‘forward fill’ and will propagate last valid observation forward.

```python
import pandas as pd
df=pd.DataFrame({"A":[5,3,None,4],
                 "B":[None,2,4,3],
                 "C":[4,3,8,5],
                 "D":[5,4,2,None]})
df
```

```python
df.ffill()
```

```python
df_spy.ffill()
```

```python
#KNN Imputer
train_knn = df_spy.copy(deep=True)
from sklearn.impute import KNNImputer
# train_knn = df_spy.copy(deep=True)

knn_imputer = KNNImputer(n_neighbors=2, weights="uniform")
train_knn['Open'] = knn_imputer.fit_transform(train_knn[['Open']])
train_knn
```

```python
#Multivariate imputation by chained equations (MICE)
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
train_mice = df_spy.copy(deep=True)

mice_imputer = IterativeImputer()
train_mice['Open'] = mice_imputer.fit_transform(train_mice[['Open']])
train_mice
```
